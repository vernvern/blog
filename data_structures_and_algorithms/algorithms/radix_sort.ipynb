# 基数排序 (radix sort)

条件：

- 数据必须可以切片
- 子排序必须是稳定排序算法
- 数据可以不等长，可以用“0”补全
- 数据位之间有递进关系
- 相对数据数量来说，数据范围不能太大，不然时间复杂度不会降到 **O(n)**


```python
from random import  Random


data = Random().choices(range(10000000000, 19999999999), k=10)
print('测试数据: %s \n' % data)

for x in range(10, -1, -1):
    data = sorted(data, key=lambda n: str(n)[x])
    print('对第%s位排序:' % x)
    for k in data:
        _k = str(k)
        print('  %s %s' % (_k[:x], _k[x:]))
    print('\n')
```
