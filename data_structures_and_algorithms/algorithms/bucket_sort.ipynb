# 桶排序 (bucket sort)

核心思想是将很大的数据分组塞到各个"桶"里面，对桶内数据排序之后，再将合并各个“桶”。

```python
from random import Random
from itertools import chain


data = Random().choices(range(100), k=100)
print('测试数据: %s \n' % data)

buckets = {x:[] for x in range(10)}
print('桶: %s' % buckets)

for x in data:
    buckets[x // 10].append(x)
print('数据装桶: %s \n' % buckets)

for k in buckets.keys():
    buckets[k] = sorted(buckets[k])
print('桶内排序: %s \n' % buckets)

result = chain(*(v for  k, v in sorted(buckets.items())))
print('合并桶: %s \n' % list(result))
```
