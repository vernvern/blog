# 队列 (queue)

特性： **先进先出**

用**数组**实现的队列叫作**顺序队列**，用**链表**实现的队列叫作**链式队列**



```python
''' 顺序队列 移动数据版'''

class ArrayQueue:
    def __init__(self, size):
        self._list = [None for x in range(size)]
        
    def enqueue(self, value):
        pass
    
    def dequeue(self):
        pass
    
    
```

```python
''' 顺序队列

一般顺序队列需要移动数据，性能较差
循环顺序队列可以解决这个问题

这种实现会有一个空位无法使用
'''

class ArrayQueue:
    def __init__(self, size):
        self._list = [None for x in range(size)]
        self.head = 0
        self.tail = 0
        self.size = size
        
    def enqueue(self, value):
        if (self.tail + 1) % self.size == self.head:
            return False
        self._list[self.tail] = value
        self.tail = (self.tail + 1) % self.size
        return True
    
    def dequeue(self):
        if self.tail == self.head:
            return False
        value = self._list[self.head]
        self._list[self.head] = None
        self.head = (self.head + 1) & self.size
        return value
    
    def __repr__(self):
        return str(self._list)
        
test = ArrayQueue(5)

test.enqueue(0)
print(test)
test.enqueue(1)
print(test)
test.enqueue(2)
print(test)
test.enqueue(3)
print(test)
test.enqueue(4)
print(test)
print(test.enqueue(5))
test.dequeue()
print(test)
test.enqueue(6)
print(test)
test.dequeue()
print(test)
test.enqueue(7)
print(test)

```

```python
''' 顺序队列

一般顺序队列需要移动数据，性能较差
循环顺序队列可以解决这个问题

这种实现不会有空位
'''

class ArrayQueue:
    def __init__(self, size):
        self._list = [None for x in range(size)]
        self.head = 0
        self.tail = 0
        self.size = size
        self.numbers = 0
        
    def enqueue(self, value):
        if self.numbers == self.size:
            return False
        self.numbers += 1
        self._list[self.tail] = value
        self.tail = (self.tail + 1) % self.size
        if self.tail == self.size  and self.numbers  !=self.size:
            self.head = 0
        return True
    
    def dequeue(self):
        print(self.tail, self.head)
        if not self.numbers:
            return False
        value = self._list[self.head]
        self._list[self.head] = None
        self.head = (self.head + 1) & self.size
        self.numbers -= 1
        return value
    
    def __repr__(self):
        return str(self._list)
        

test = ArrayQueue(5)

test.enqueue(0)
print(test)
test.enqueue(1)
print(test)
test.enqueue(2)
print(test)
test.enqueue(3)
print(test)
test.enqueue(4)
print(test)
print(test.enqueue(5))
test.dequeue()
print(test)
test.enqueue(6)
print(test)
test.dequeue()
print(test)
test.enqueue(7)
print(test)

```
