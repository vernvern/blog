# 正则表达式

## 简单的正则匹配

要求:

- ***** : 匹配任意多个(0或更多)任意字符
- **?** : 匹配零或一个任意字符


```python
class Regex:
    def __init__(self, regex):
        self.regex = regex
        self.regex_len = len(regex)
        
    def match(self, text):
        self.matched = False
        self.rmatch(0, text, 0, len(text))
        print('result: %s' % self.matched)
        return self.matched
    
    def rmatch(self, regex_index, text, index, text_len):
        print(regex_index, self.regex_len, self.regex, text, index, text_len)

        if self.matched:
            return
        
        if regex_index == self.regex_len or index == text_len:
            if regex_index == self.regex_len and index == text_len:
                self.matched = True
            return
        regex = self.regex
        
        debug_msg = 'regex[%s]: %s' % (regex_index, regex[regex_index])
        debug_msg += '  text[%s]: %s'%(index, text[index])
        print(debug_msg)        

        if regex[regex_index] == '*':
            for t_index in range(index, text_len+1):
                self.rmatch(regex_index+1, text, t_index, text_len)
        elif regex[regex_index] == '?':
            self.rmatch(regex_index+1, text, index, text_len)
            self.rmatch(regex_index+1, text, index+1, text_len)
        elif regex[regex_index] == text[index]:
            self.rmatch(regex_index+1, text, index+1, text_len)
            
        
test = Regex('a?')
test.match('ab')
print('-'*30)
     
test = Regex('?b')
test.match('ab')
print('-'*30)

test = Regex('a*')
test.match('ab')
print('-'*30)

test = Regex('*b')
test.match('acb')
print('-'*30)

test = Regex('*c')
test.match('acb')
print('-'*30)
```
